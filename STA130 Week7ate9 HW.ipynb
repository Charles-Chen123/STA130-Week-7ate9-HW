The Simple Linear Regression (SLR) model is a statistical tool used to understand the relationship between two variables: a predictor (independent variable)  X  and an outcome (dependent variable)  Y . In its theoretical form, the model assumes that the outcome  Y  can be predicted by a linear combination of  X , where the slope coefficient ( \beta_1 ) represents the rate of change in  Y  for a unit increase in  X , and the intercept ( \beta_0 ) represents the value of  Y  when  X  is zero. However, real-world data often deviates slightly from the exact line due to random variability, so an error term  is added to account for this “noise.”

In the SLR model, we assume that the error term  \epsilon  follows a normal distribution, making  Y  also normally distributed for a given  X . This model can generate data points that fluctuate around the line defined by  \beta_0 + \beta_1 X , with the spread of these points influenced by the error term’s standard deviation  \sigma .

```python
1. 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
n = 100  # Number of samples
beta0 = 2  # Intercept
beta1 = 0.5  # Slope
sigma = 1.0  # Standard deviation of error term

# Generate predictor values X
np.random.seed(0)  # For reproducibility
X = np.random.uniform(0, 10, n)

# Generate error term from normal distribution
epsilon = norm.rvs(loc=0, scale=sigma, size=n)

# Compute the outcome variable Y
Y = beta0 + beta1 * X + epsilon

# Plotting the theoretical line and the simulated data points
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color="blue", label="Simulated data points")
plt.plot(X, beta0 + beta1 * X, color="red", label="Theoretical line (no error)")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.legend()
plt.title("Simple Linear Regression Simulation")
fig.show(renderer="png")
```

```python
2. 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Parameters for the simulation
n = 100  # Number of samples
beta0 = 2  # Intercept
beta1 = 0.5  # Slope
sigma = 1.0  # Standard deviation of error term

# Generate predictor values X
np.random.seed(0)  # For reproducibility
X = np.random.uniform(0, 10, n)

# Generate error term from normal distribution
epsilon = np.random.normal(0, sigma, n)

# Compute the outcome variable Y
Y = beta0 + beta1 * X + epsilon

# Create a DataFrame for the simulated data
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols('Y ~ X', data=data).fit()

# Print the summary of the fitted model
print(model.summary())

# Visualization of the fitted model and data
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color="blue", label="Simulated data points")
plt.plot(data['X'], model.fittedvalues, color="red", label="Fitted regression line")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.legend()
plt.title("Fitted Simple Linear Regression Model")
fig.show(renderer="png")
```

The line which is chosen for the fitted model in ordinary least squares is the line that minimum the sum of the squared residuals. residuals are the vertical distances between the observed data points and the line. 

The reason why squaring the residuals is to make sure that all deviations are positive and to amplify the impact of larger error on the overall loss function that OLS method tries to minimize. Without squaring, positive residuals may be cancelled out by negative residuals, leading to misleading fit. The OLS method aims to make sure the line is as close as possible to the data points through minimizing the sum of squared residuals to find the the best linear approximation of the relationship between x and Y.

```python
10.

import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
from scipy import stats

# Sample data similar to Old Faithful if the original dataset is not available
data = {
    "waiting": [80, 54, 74, 62, 85, 55, 88, 95, 68, 75, 80, 56, 60, 82, 74] * 11,  # 165 samples to mimic original size
    "duration": [4.2, 2.5, 3.8, 2.7, 4.5, 2.6, 4.8, 5.1, 2.9, 3.9, 4.1, 2.4, 2.6, 4.3, 3.7] * 11
}
old_faithful = pd.DataFrame(data)

# Define the threshold for "long wait times"
long_wait_limit = 71
long_wait = old_faithful['waiting'] > long_wait_limit
n_bootstraps = 1000  # Number of bootstrap samples

# Step 1: Bootstrap samples and fit models to collect slope coefficients
bootstrapped_slope_coefficients = []

for _ in range(n_bootstraps):
    boot_sample = old_faithful[long_wait].sample(n=long_wait.sum(), replace=True)
    model = smf.ols('duration ~ waiting', data=boot_sample).fit()
    bootstrapped_slope_coefficients.append(model.params.iloc[1])  # Collect slope coefficient with iloc

# Visualize the bootstrapped slope coefficients
fig = px.histogram(bootstrapped_slope_coefficients, nbins=30, title="Bootstrapped Slope Coefficients")
fig.show()

# Step 2: Simulate samples under the null hypothesis (slope = 0)
old_faithful_simulation = old_faithful[long_wait].copy()
simulated_slope_coefficients = []

for _ in range(n_bootstraps):
    # Generate simulated durations under the null hypothesis of no association
    old_faithful_simulation['duration'] = 1.65 + 0 * old_faithful_simulation['waiting'] + stats.norm(loc=0, scale=0.37).rvs(size=long_wait.sum())
    model = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    simulated_slope_coefficients.append(model.params.iloc[1])  # Collect slope coefficient under null with iloc

# Visualize the simulated slope coefficients under the null hypothesis
fig = px.histogram(simulated_slope_coefficients, nbins=30, title="Simulated Slope Coefficients under Null Hypothesis")
fig.show()

# Step 3: Calculate 95% confidence interval for bootstrapped slope coefficients
ci_bootstrap = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])
print("95% Bootstrapped Confidence Interval for Slope:", ci_bootstrap)

# Step 4: Check if 0 is within the confidence interval
contains_zero = ci_bootstrap[0] <= 0 <= ci_bootstrap[1]
print("Does the confidence interval contain 0?", contains_zero)

# Step 5: Calculate the observed slope coefficient and the simulated p-value
observed_model = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit()
observed_slope = observed_model.params.iloc[1]  # Use iloc for observed slope

# Calculate the p-value based on simulated null distribution
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()
print("Simulated p-value:", simulated_p_value)
```
